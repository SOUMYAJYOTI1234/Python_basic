## map operator
syntax: map(function, iterable)

```python
seq=[1,2,3,4,5,6]
list(map(lambda var: var*2,seq))
```

```python
seq=[1,2,3,4,5,6]
def times2(seq):
    times2_values=[]
    for num in seq:
        times2_values.append(num*2)
    return times2_values
times2(seq)
```

```python
# map(function, iterable) --> syntax of map
# map() applies the given function to each element of the iterable (seq here).
# So lambda var: var * 2 runs on each number in seq.
```

```python
# Transform a list of strings so that each character in each string becomes an individual list item using map function.

L=["apple","banana"]
list(map(list,L))
```

Real-Life Example of pass
Scenario: Planning a feature for later

You're designing a mobile app, and you know you will add the notifications feature, but you haven't written the code yet.
So you leave a placeholder.

```python
features = ["login", "profile", "notifications", "payments"]

for f in features:
    if f == "notifications":
        pass  # You will implement it later
    print("Feature ready:", f)
```

continue Statement


Used to skip the current iteration and move to the next one.


Example: Skip printing the number 5

```python
for i in range(1, 10):
    if i == 5:
        continue
    print(i)
```

Real-Life Example of continue


Scenario: Ignore damaged items while processing orders


Suppose you process an order list but skip the ones labeled "damaged".

```python
orders = ["order1", "damaged", "order2", "order3", "damaged", "order4"]

for order in orders:
    if order == "damaged":
        print("Skipping damaged item.")
        continue
    print("Processing:", order)
```

break Statement


Used to exit the loop completely.


Example: Stop when number becomes 5

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

Real-Life Example of break


Scenario: Stopping when you find the item you are searching for


Imagine you are searching for your ID card in a list of items.
Once you find it, you stop searching.

```python
items = ["keys", "wallet", "pen", "ID card", "notebook"]

for item in items:
    if item == "ID card":
        print("Found ID card! Stopping search.")
        break
    print("Checking:", item)
```

```python
def fact(n):
    if(n==0 or n==1):
        return 1
    else:
        return n*fact(n-1)
fact(5)
```

```python
def calc_sum(n):
    if(n==0):
        return 0 
    return calc_sum(n-1) + n
calc_sum(3)
```

```python
def print_list(list,idx=0):
    if(idx==len(list)):
        return
    print(list[idx])
    print_list(list,idx+1)
fruits=["mango","litchi","apple","banana"]
print_list(fruits)
```

```python
# map function
def cube(x):
    return x*x*x
l=[1,2,3,4]
newl=[]
for i in l:
    newl.append(cube(i))
print(newl)
```

```python
def cube(x):
    return x*x*x
l=[1,2,3,4]
newl= list(map(cube,l))
print(newl)
```

```python
# map using lambda function
l=[1,3,4,5,6,7]
newl=list(map(lambda x: x*x*x, l))
print(newl)
```

```python
# filter function
l=[1,2,4,6,4,3]
def filter_function(a):
    return a>2
newl=list(filter(filter_function,l))
print(newl)
```

```python
# reduce function
from functools import reduce
numbers =[1,2,3,4,5]
def mysum(x,y):
    return x+y
sum=reduce(mysum,numbers)
print(sum)
```

```python
class Atm:
    # constructor(special function)-> superpower->after creating an object of the class, this function will be called automatically
    def __init__(self):
        self.pin=''
        self.balance=0
        self.menu()
    def menu(self):
        user_input = input("""
              Hi, how can I help you?
              1. Press 1 to create pin
              2. Press 2 to change pin
              3. Press 3 to check balance
              4. Press 4 to withdraw cash
              5. Anything else to exit
              """)
        if user_input == '1':
            # create pin
            self.create_pin()
        elif user_input == '2':
            # change pin
            self.change_pin()
        elif user_input == '3':
            # check balance
            self.check_balance()
        elif user_input == '4':
            # withdraw cash
            self.withdraw_cash()
        else:
            print("Thank you for using our services.")
            exit()
    def create_pin(self):
        user_pin = input("enter your pin: ")
        self.pin = user_pin
        user_balance = int(input('enter balance:'))
        self.balance = user_balance

        print("Pin created successfully")
        self.menu()

    
    def change_pin(self):
        old_pin = input("enter old pin: ")
        if old_pin == self.pin:
            new_pin = input("enter new pin: ")
            self.pin = new_pin
            print("Pin changed successfully")
            self.menu()
        else:
            print("Incorrect pin")
            self.menu()

    def check_balance(self):
        user_pin = input("enter your pin: ")
        if user_pin == self.pin:
            print("Your balance is: ", self.balance)
            self.menu()
        else:
            print("Incorrect pin")
            self.menu()

    def withdraw_cash(self):
        user_pin = input("enter your pin: ")
        if user_pin == self.pin:
            amount = int(input("enter amount to withdraw: "))
            if amount <= self.balance:
                self.balance = self.balance-amount
                print("Please collect your cash")
                print("Your new balance is: ", self.balance)
                self.menu()
            else:
                print("Insufficient balance")
                self.menu()
        else:
            print("Incorrect pin")
            self.menu()
```
