# scatter Plots
##### The Scatter Plot is a mainstay of statistical vizualization. it depicts the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about  whether there is any meaningful relationship between them.

#### Although, Scatter plot can be plotted for both numerical as well as categorical data, it's mostly used to depict relationship between numerical data. 

#### There are 2 ways to draw a sctterplot in seaborn.
1. relplot(also known as relation plot, it's a figure level function)
2. scatterplot(Axes Level function)

#### As we will see, these visualisations can be quite helpful because they use simple and easily understandable representations of data even on complex data by mapping up to three additional variables using the semantics of hue, size, and style.

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

## Plot a scatter plot between total_bill and tip

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="seaborn.axisgrid")
```

```python
# relplot
sns.relplot(x='total_bill', y='tip', data=tips)
plt.show()
```

## The hue parameter

```python
sns.relplot(x='total_bill', y='tip',hue='smoker', data=tips)
# The semantics 'hue' corresponds to a categorical data
```

## The style parameter

```python
sns.relplot(x='total_bill', y='tip',hue='smoker',style='sex', data=tips)
# The semantics 'style' corresponds to a categorical data
```

## The size parameter

```python
sns.relplot(x='total_bill', y='tip',hue='smoker',style='sex',size='size',sizes=(10,100), data=tips)
# The semantics 'size' corresponds to a numerical data
```

# Plotting with categorical Data
### Catplot(Categorical plot)(Figure level function)

## 1. Categorical Scatterplots
### a. Stripplot
### b. Swarmplot

```python
# The main criteria of a Categorical Scatterplot is,
# x-axis --> Categorical data
# Y-axis --> Numerical data
```

```python
tips.head()
```

```python
# Stripplot
sns.catplot(x='day', y='tip',kind='strip', data=tips )
```

```python
# Swarmplot
sns.catplot(x='day', y='tip', kind='swarm', data=tips)
```

```python
sns.catplot(x='day', y='tip', kind='swarm', hue='sex',data=tips)
```

```python
# we only have semantic hue variable in catplot, we don't have the semantic
# style or size variable in catplot
```

# Boxplots
#### A boxplot is a standardized way of displaying the distribution of data based on a five number summary(minimum, first quartile(Q1), median, third quartile(Q3), and maximum).it can tell you about your outliers and what their values are. it can also tell you if your data is symmetrical, how tightly your data is groupped  and if and and how your data is skewed.

```python
sns.boxplot(tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='box',data=tips)
```

```python
sns.catplot(x='day', y='total_bill',hue='sex', kind='box',data=tips)
```

# violin Plots
#### Violin plots are a method of plotting numeric data and can be considered a combination of the box plot with a Probability density function.

```python
# Violin plot is symmetrical in both sides so we don't have to observe
# both the sides.
```

#### The advantage of the violin plot over the box plot is that aside showing the above mentioned statistics it also shows the entire distribution of the data.

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill',kind='violin',data=tips)
```

```python
sns.catplot(x='day', y='total_bill',hue='sex',kind='violin',data=tips)
```

```python
sns.catplot(x='day', y='total_bill',hue='sex',kind='violin',split=True, data=tips)
```

# Barplot and Countplot

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=tips)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex',kind='bar',data=tips)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex',estimator=np.median,kind='bar',data=tips)
```

```python
sns.catplot(x='sex',kind='count',data=tips)
# countplot gives the count of a categorical variable
```

```python
sns.catplot(x='sex',kind='count',hue='smoker',data=tips)
```

# Heatmaps
#### A heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors.

```python
sns.load_dataset('flights')
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year', columns='month', values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, linewidth=1.0)
```

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x, cbar=False, linewidth=0.5, annot=True, fmt='d' )
```

```python
 sns.heatmap(x, cbar=False, linewidth=0.5, annot=True, fmt='d', cmap='viridis')
```

# Clustermap

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True, metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True, row_cluster=False, metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True, col_cluster=False, metric='correlation')
```

# Jointplot
#### Jointplot creates a multi-panel figure that shows both the bivariate(or joint)
#### relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.
#### We can use jointplots in the following 4 manners: 
#### 1.Scatterplots
#### 2.Hexbin plots
#### 3.Kdeplots
#### 4.Regplot

```python
tips=sns.load_dataset('tips')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex',data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde',data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=tips)
```

# Distplots
#### Distplots are used to do univariate analysis on a continuous/numerical data.
#### Distplots combines 3 things in one graph:
#### 1.Histogram
#### 2.Kernel Density Estimation(PDF)
#### 3.Rugplot

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean())
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

### Plotting a Distplot 

```python
sns.distplot(titanic['age'])
```

```python
sns.displot(titanic['age'],kde=False)
```

```python
sns.distplot(titanic['age'], hist=False)
```

# Plotting Rugplot

```python
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

### Use of Distplot

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
